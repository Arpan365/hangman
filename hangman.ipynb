```python
import random
import hangman_words
import hangman_art
import os
```

```python
word_list=hangman_words.word_list
stages=hangman_art.stages
chosen_word=random.choice(word_list)
```

```python
os.system('cls')
print(hangman_art.logo)
display=[]
for _ in range(len(chosen_word)):
    display+="_"
print(display,end="")

lives=6
while display!=list(chosen_word):
    guess=input("Guess a letter: ").lower()
    if guess not in list(chosen_word):
        lives-=1
        print(f"You have guessed {guess} and you have lose a life, {lives} lives left")
        print(stages[lives])
        if lives==0:
            break
    else:
        
        for position in range(len(chosen_word)):
            letter=chosen_word[position]
            if letter==guess:
                display[position]=letter
                if guess==list(display):
                    print("You Have already guessed this letter")
        
    print(display)
if lives==0:
    print("GAME OVER")
    display!=list(chosen_word)
if display==list(chosen_word):
    print('YOU WIN')
```
